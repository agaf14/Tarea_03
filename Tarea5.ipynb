```python
#Código y ejemplo de la solución

#Librerias utilizadas
import matplotlib.pyplot as plt
import numpy as np
import math
import ipywidgets as widgets
from ipywidgets import interact

#Código de la solución y funciones

def q(T):
  return 1*20*((25-T)/0.3)

q2= lambda x: 0.75*20*((25-T)/0.3)
q3= lambda x: 1.25*20*((25-T)/0.3)

T=np.linspace(-15,38,15+38)

plt.plot(T,q(T),label='1W/mK')
plt.plot(T,q2(T),label='0.75W/mK')
plt.plot(T,q3(T),label='1.25W/mK')
plt.xlabel("Temperatura")
plt.ylabel("Calor")
plt.legend()
plt.grid()
```
